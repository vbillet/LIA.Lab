```python
import yaml
```

```python
doc = """
Library:
    Description: Pascal généraliste
    Contexte: Delphi
    Auteur: Vincent BILLET
    Version: 1.0
Commands:
    - Command:
        Name: AddUnit
        Question: Comment créer une unité ayant pour nom [MASK] ?
        Params:
            - Param:
                Nom: NomUnite
                Libelle: nom de l'unité
                Traitements: Variabilize
                Questions:
                  - Quel est le nom de l'unité ?
        Ou: Contenu
        Pourquoi: Pour contenir le code source.
        Template:
            Unit §NomUnite§
            
            interface
            
            uses
            [UsesListInt]
            
            const
            [ConstList]
            
            type
            [TypeList]
            
            var
            [GlobalVarList]
            
            implementation
            
            uses
            [UsesListImp]
            
            [Implementation]
            
            end.
    - Command:
        Name: addConst
        Question: Comment ajouter une constante [MASK] ayant pour valeur [MASK] ?
        Params:
            - Param: 
                Nom: NomConstante
                Libelle: nom de la constante
                Traitements: Variabilize
                Questions:
                   - Quel est le nom de la constante ayant pour valeur §Valeur§ ?
            - Param: 
                Pseudo: Valeur
                Nom: valeur de la constante
                Traitements: TestString
                Questions:
                   - Quelle est la valeur de la constante §NomConstante§ ?
        Ou: ConstList
        Pourquoi: §NomConstante§ ne change jamais, alors on le défini en constante avec pour valeur |Valeur|
        Template:
            §*§C§NomConstante§ = §Valeur§;
    - Command:
        Name: addClass
        Question: Comment ajouter une classe [MASK] ?
        Params:
            - Param:
                Nom: NomClasse
                Libelle: nom de la classe
                Traitements: Variabilize
                Questions:
                  - Quel est le nom de la classe ?
        Ou: TypeList
        Pourquoi: Permet de regrouper des données et les méthodes qui leurs sont associées.
        Template:
            §*§TLp§NomClasse§ = class;
            §*§§*§private
            [private]
            §*§§*§protected
            [protected]
            §*§§*§public
            [public]
            end;
    - Command:
        Name: addProcedure
        Question: Comment ajouter une procédure [MASK] ?
        Params:
            - Param:
                Nom: NomProcedure
                Libelle: nom de la procedure
                Traitements: Variabilize
                Questions:
                  - Quel est le nom de la procedure ?
        Ou: implementation
        Pourquoi: Permet d'effectuer une liste de traitements.
        Template:
            procedure §NomProcedure§([Parametres]);
            const
            [Constantes]
            var
            [variables]
            begin
              [instructions]
            end;
    - Command:
        Name: addfunction
        Question: Comment ajouter une fonction [MASK] qui retourne [MASK] ?
        Params:
            - Param:
                Nom: NomFonction
                Libelle: nom de la fonction
                Traitements: Variabilize
                Questions:
                  - Quel est le nom de la fonction ?
            - Param:
                Nom: TypeRetour
                Libelle: Type de retour
                Traitements: Variabilize
                Questions:
                  - Quel est le type de retour de la fonction ?

        Ou: implementation
        Pourquoi: Permet d'effectuer une liste de traitements.
        Template:
            function §NomFonction§([Parametres]) &dp; §TypeRetour§ ;
            const
            [Constantes]
            var
            [variables]
            begin
              [instructions]
            end;
"""
```

```python
d=yaml.safe_load(doc)
for c in d["Commands"]:
    print(c["Command"]["Name"], ":", c["Command"]["Question"], len(c["Command"]["Params"]),"paramètres")
```
